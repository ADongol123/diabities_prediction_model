```python
# Define BMI categories based on WHO classification
def classify_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal weight'
    elif 25 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

# Apply classification to the BMI column
dataset['BMI_Category'] = dataset['BMI'].apply(classify_bmi)

# Count the number of people in each BMI category
bmi_counts = dataset['BMI_Category'].value_counts()
print(bmi_counts.index)
# Plotting the counts as a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(hue=bmi_counts.index, y=bmi_counts.values, palette="Set2")
plt.title("Distribution of BMI Categories")
plt.xlabel("BMI Category")
plt.ylabel("Number of People")
plt.show()

plt.figure(figsize=(7, 7))
explode = (0.1, 0) 
plt.pie(bmi_counts.values, labels=bmi_counts.index, autopct='%1.1f%%', colors=sns.color_palette("Set2"), shadow=True, startangle=90)
plt.title("Proportion of People by BMI Category")
plt.show()
```

# Creating a model

```python
dataset= pd.read_csv("../datasets/diabetes.csv")
dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from helper_functions import accuracy_fn
```

```python
X=dataset.drop('Outcome',axis=1).values### independent features
y=dataset['Outcome'].values###dependent features
```

```python
# data = pd.get_dummies(dataset, columns=['BMI_Category'], drop_first=True)
```

```python
# X = dataset.drop('Outcome', axis=1).values  # Features
# y = dataset['Outcome'].values  # Target variable
# X,y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
len(X_test),len(X_train),len(y_train),len(y_test)
```

```python
dataset.shape
```

```python
X_train = torch.FloatTensor(X_train)
y_train = torch.LongTensor(y_train)
X_test = torch.FloatTensor(X_test)
y_test = torch.LongTensor(y_test)

X_test.dtype,X_train.dtype,y_train.dtype,y_test.dtype
```

```python
# print(data.dtypes)  # Check data types of all columns
```

```python
class Diabities_Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.first = nn.Linear(in_features=8, out_features=20)
        # self.batch_norm1 = nn.BatchNorm1d(20)
        # self.dropout1 = nn.Dropout(0.2)
        
        self.second = nn.Linear(in_features=20, out_features=20)
        # self.batch_norm2 = nn.BatchNorm1d(20)
        # self.dropout2 = nn.Dropout(0.2)
        
        self.out = nn.Linear(in_features=20, out_features=2)

    def forward(self,x):
        x=F.relu(self.first(x))
        # x = self.dropout1(x)
        
        x=F.relu(self.second(x))
        # x=self.dropout2(x)
        
        x=self.out(x)
        return x
```

```python
# Creating a model
torch.manual_seed(42)
model_0 = Diabities_Model()
model_0.parameters
```

```python
# Creating a loss function and optimizer
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_0.parameters(),lr=0.01)
```

```python
# Traning and Testing the model
y_train = y_train.long()
y_test = y_test.long()

torch.manual_seed(42)

epochs = 500

final_losses = []

for epoch in range(epochs):
    model_0.train()

    y_pred=model_0(X_train)

    loss= loss_fn(y_pred,y_train)

    acc = accuracy_fn(y_true=y_train,
                     y_pred = y_pred.argmax(dim=1))
    
    final_losses.append(loss.detach().item())
    
    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

    # Testing the model

    model_0.eval()
    with torch.inference_mode():
        test_pred = model_0(X_test)
        test_loss = loss_fn(test_pred,y_test)
        test_acc = accuracy_fn(y_true=y_test,y_pred=test_pred.argmax(dim=1))
    if epoch % 10 == 0:
        print(f"The epoch is: {epoch} | loss: {test_loss} | Test Accuracy: {test_acc:.4f}% ")


test_loss
```

```python
plt.plot(range(epochs), final_losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```
